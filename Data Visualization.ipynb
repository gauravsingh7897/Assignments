#### Problem: 
In this assignment students have to transform iris data into 3 dimensions
and plot a 3d chart with transformed dimensions and colour each data
point with specific class.

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn import decomposition
from sklearn import datasets
%matplotlib inline
```

```python
dataset = datasets.load_iris()
y = dataset.target
```

```python
x_reduced_dim = decomposition.PCA(n_components=3).fit_transform(dataset.data)
print("Before applying PCA iris dataset shape:",dataset.data.shape)
print("After applying PCA iris dataset shape:",x_reduced_dim.shape)
print("3D Chart:")
fig = plt.figure(1,figsize=(10,6))
ax = Axes3D(fig,elev=-150,azim=105)
ax.scatter(x_reduced_dim[:,0],x_reduced_dim[:,1],x_reduced_dim[:,2],c=y,cmap=plt.cm.viridis)
plt.title('PCA in 3 Dim')
ax.set_xlabel('1st Eigen Vector')
ax.set_ylabel('2nd Eigen Vector')
ax.set_zlabel('3rd Eigen Vector')
plt.show()
```
